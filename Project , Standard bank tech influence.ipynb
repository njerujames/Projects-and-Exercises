# Load Datasets

```python
training_data = pd.read_csv('D:/ML/Project,Standard bank tech influence/Train.csv')
testing_data = pd.read_csv('D:/ML/Project,Standard bank tech influence/Test.csv')
training_data.tail()
```

# Data Wrangling

creating two new features based on time

```python
# load period in days
LoanPeriod = pd.Series(delta.days for delta in(pd.to_datetime(training_data['DueDate']) - pd.to_datetime(training_data['IssuedDateLoan'])))
```

```python
# Loan payment Period in day
PaidTime =  pd.Series(delta.days for delta in(pd.to_datetime(training_data['PaidOnDate']) - pd.to_datetime(training_data['IssuedDateLoan'])))
```

```python
# Adding the new features in the dataset
training_data.insert(25,'LoanPeriod',LoanPeriod.values)
training_data.insert(26,'PaidTime',PaidTime.values)
training_data.head()
```

```python
training_data.shape
```

Handling Missing Values, That is Customers who did not apply for a Loan

```python
training_data.dropna(inplace = True)
```

Handling Categorical data

```python
label_encoder = LabelEncoder()
```

```python
# Categorical boolean mask
categorical_feature_mask = training_data.dtypes == object
categorical_cols = training_data.columns[categorical_feature_mask].tolist()
```

```python
categorical_cols
```

```python
training_data[categorical_cols] = training_data[categorical_cols].apply(lambda col: label_encoder.fit_transform(col))
```

```python
training_data.head()
```

```python
Feature_data = training_data.iloc[:,:-1].values
Target_data = training_data.iloc[:,-1].values
Feature_data.shape
```

```python
Target_data.shape
```

```python
# Standardize the feature matrix
scaler = StandardScaler()
Feature_data = scaler.fit_transform(Feature_data)
Feature_data
```

Dimensionality reduction uding PCA

```python
# Create a PCA that will retain 99% of variance
pca = PCA(n_components=0.99, whiten=True)
# Conduct PCA
Feature_pca = pca.fit_transform(Feature_data)
print(Feature_data.shape)
Feature_pca.shape
```

```python
Feature_pca[1]
```

```python
Feature_data[1]
```

# Model Selection

Selecting Best Models Using Randomized Search

```python
# Create logistic regression
logistic = linear_model.LogisticRegression()
# Create range of candidate regularization penalty hyperparameter values
penalty = ['l1', 'l2']
# Create distribution of candidate regularization hyperparameter values
C = uniform(loc=0, scale=4)
# Create hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)
# Create randomized search
randomizedsearch = RandomizedSearchCV(
logistic, hyperparameters, random_state=1, n_iter=100, cv=5, verbose=0,
n_jobs=-1)
# Fit randomized search
best_model = randomizedsearch.fit(Feature_pca, Target_data)
```

```python
best_model
```

```python
# Define a uniform distribution between 0 and 4, sample 10 values
uniform(loc=0, scale=4).rvs(10)
```

```python
# View best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```python
# Predict target vector
best_model.predict(Feature_pca)
```

Selecting Best Models Using Exhaustive Search

```python
# Create range of candidate regularization hyperparameter values
C = np.logspace(0, 4, 10)
# Create dictionary hyperparameter candidates
hyperparameters = dict(C=C, penalty=penalty)
# Create grid search
gridsearch = GridSearchCV(logistic, hyperparameters, cv=5, verbose=0)
# Fit grid search
best_model = gridsearch.fit(Feature_pca, Target_data)
```

```python
# View best hyperparameters
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
np.logspace(0, 4, 10)
```

```python
# Predict target vector
best_model.predict(Feature_pca)
```

# Model Evaluation

Cross-Validating Models

```python
# Create k-Fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=1)
```

```python
# Create a pipeline that standardizes, then runs logistic regression
pipeline = make_pipeline(scaler, logistic)
```

```python
# Conduct k-fold cross-validation
cv_results = cross_val_score(pipeline, # Pipeline
Feature_pca, # Feature matrix
Target_data, # Target vector
cv=kf, # Cross-validation technique
scoring="accuracy", # Loss function
n_jobs=-1) # Use all CPU scores
```

```python
# Calculate mean
cv_results.mean()
```
