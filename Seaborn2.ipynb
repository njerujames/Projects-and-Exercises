# Loading Dependancies

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

# Figure Aesthetic

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

sinplot()
plt.show()
```

```python
# To change the same plot to Seaborn defaults, use the set() function

def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

sb.set()
sinplot()
plt.show()
```
